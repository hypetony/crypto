```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Define the API endpoint and parameters
url = "https://api.coingecko.com/api/v3/coins/bitcoin/market_chart"
params = {
    'vs_currency': 'usd',
    'days': '30',  # Get the last 30 days of data
    'interval': 'daily'
}

# Send the GET request
response = requests.get(url, params=params)
data = response.json()

# Extract and organize the data
prices = data['prices']
dates = [datetime.utcfromtimestamp(item[0] / 1000).strftime('%Y-%m-%d') for item in prices]
price_values = [item[1] for item in prices]

# Create a DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Price (USD)': price_values
})

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price (USD)'], marker='o', color='b', label='Bitcoin Price')
plt.xticks(rotation=45)
plt.title('Bitcoin Price Trend (Last 30 Days)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()

# Display the DataFrame
print(df)
```

# Bitcoin Price Trend Analysis (Last 30 Days)

## Objective:
This analysis aims to visualize the price trend of Bitcoin (BTC) over the past 30 days using publicly available market data from the CoinGecko API. The dataset provides daily price values for Bitcoin, which are organized by date.

## Data Collection:
The data is collected via the CoinGecko API using the endpoint for Bitcoin's market chart in USD over the past 30 days. Each data point represents the Bitcoin price on a given day.

## Key Insights:
Price Fluctuations: Over the past 30 days, Bitcoin has shown notable fluctuations in its price, which can be observed through its daily variations.
Trend: The graph visually represents Bitcoins price fluctuations, helping to identify any patterns, peaks, or dips during the given period.
Visualization:
A line plot is provided to give a clear picture of the changes in Bitcoin's price over time. The x-axis represents the dates, while the y-axis represents the price in USD. This visual helps in identifying trends, volatility, and any potential correlations to external factors.

## Conclusion:
This analysis provides a snapshot of Bitcoin's price volatility over the last 30 days, offering insights into its short-term market behavior.
